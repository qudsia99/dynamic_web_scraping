```python
# dependencies 
from bs4 import BeautifulSoup 
import pandas as pd
from splinter import Browser
```

```python
# Configure browser
browser = Browser('chrome')
```

```python
# Grabbing URL
url = 'https://mississauga.bibliocommons.com/list/share/2423508199/2641596647'
browser.visit(url)
```

```python
# if response.status_code == 200:
#     print('Success!')
# else:
#     print('Failed!')
```

```python
# Creating a soup object
html = browser.html
soup = BeautifulSoup(html,'html.parser')
```

```python
# Finding the correct element with data, utilized xpath on inspect element
data = soup.find_all('div', class_='list_item_info')
```

```python
# double check
print(data)
```

```python
# Create lists and fetch each value needed within the larger element.
# then create a for loop.

book_name = []
book_subtitle = []
author = []
release_year = []

for book in data:
    book_name.append(book.find('div', class_ = 'list_item_title').find('a').get_text(strip=True))
    subtitle = book.find('div', class_ = 'list_item_subtitle')
    if subtitle:
        book_subtitle.append(subtitle.get_text(strip=True))
    else:
        book_subtitle.append('N/A')
    author.append(book.find('div', class_ = 'list_item_author').find('a').get_text(strip=True))
    year = (book.find('div', class_ = 'list_item_format').get_text(strip=True))
    year = year.split('-')[-1]
    release_year.append(year)
    
print(book_name)
print(book_subtitle)
print(author)
print(release_year)

```

```python
# Convert it to a dataframe using a dictionary
df = pd.DataFrame({'Title':book_name, 'Subtitle':book_subtitle, 'Author':author, 'Release Year':release_year})
```

```python
# Exporting to CSV
df.to_csv('book_catalogue.csv', index = False)
```
